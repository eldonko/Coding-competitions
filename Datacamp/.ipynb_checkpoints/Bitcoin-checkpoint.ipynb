# Should our fund invest in Bitcoin?

## ðŸ“– Background
In this report we will analyze how Bitcoin performs as an asset and if it should be added to a portfolio. Comparisons are made with the S&P500 and Gold as well as the inflation in the US. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
import math
from sklearn.linear_model import LinearRegression

plt.rcParams["figure.figsize"] = (10, 7)
```

```python
bitcoin = pd.read_csv('./data/bitcoin-usd.csv', parse_dates=['date'], index_col='date')
sp500 = pd.read_csv('./data/sp500.csv', parse_dates=['date'], index_col='date')
monthly_data = pd.read_csv('./data/monthly_data.csv', parse_dates=['date'], index_col='date')
```

```python
color_bitcoin = 'darkorange'
color_sp500 = 'navy'
color_gold = 'darkgoldenrod'
```

# Returns
With Bitcoin, very high returns are possible. Daily changes of around 10% both up and down are not an exception, often times it even reaches 20%. In early 2020, Bitcoin reached its highest daily loss when it was almost 40% down. 

The monthly and annual returns are even more extraordinary. Espacially the two peaks in late 2017 and early 2021 gave big monthly and annual returns. At the peak of early 2021, Bitcoin was up around 1000% compared to the previous year. In 2017, this number was even higher when the annual return reached more than 2000%. 

```python
bitcoin['daily_return'] = bitcoin['close'].pct_change(1)
bitcoin['monthly_return'] = bitcoin['close'].pct_change(30)
bitcoin['annual_return'] = bitcoin['close'].pct_change(360)

bitcoin[['close', 'daily_return', 'monthly_return', 'annual_return']].plot(subplots=True, figsize=(10, 7), grid=True, title='Changes of Bitcoin close price daily, monthly and annually', xlabel='Date', legend=True)
```

# Performance
## Overall performance
Over the whole period since 2014, Bitcoin outperformed both the S&P500 and Gold dramatically. Until 2017, the three assets seem to be on one level, but since then, Bitcoin started to grow exponentially. At the end of the period, the normalized value of Bitcoin is around 56 times the one of the S&P500 and almost 92 times the one of Gold. 

When considering an **initial investment of \$1000** in all of the 3 assets, the money invested in Bitcoin would now be worth **\$131798**. Compared to this, the **\$2339** and **\$1438** for the S&P500 and Gold seem neglectable. 

```python
start_investment = 1000
normalized = pd.DataFrame()

# Bitcoin
normalized['bitcoin'] = bitcoin['close'].div(bitcoin['close'].iloc[0]).mul(start_investment)

# S&P500
normalized['sp500'] = sp500['close'].div(sp500['close'].iloc[0]).mul(start_investment)
normalized['sp500'].fillna(method='ffill', inplace=True)

# Gold
normalized['gold'] = monthly_data['gold_usd'].div(monthly_data['gold_usd'].iloc[0]).mul(start_investment)
normalized['gold'].fillna(method='ffill', inplace=True)

# Plot
normalized.plot(grid=True, title='Performance of Bitcoin, S&P500 & Gold over the whole period', color=[color_bitcoin, color_sp500, color_gold])

# End values
print('End ratios:')
print('Bitcoin/S&P500: ', normalized['bitcoin'].iloc[-1]/normalized['sp500'].iloc[-1])
print('Bitcoin/Gold: ', normalized['bitcoin'].iloc[-1]/normalized['gold'].iloc[-1])
print('\nEnd values:')
print('Bitcoin: ', normalized['bitcoin'].iloc[-1])
print('S&P500: ', normalized['sp500'].iloc[-1])
print('Gold: ', normalized['gold'].iloc[-1])
```

## Monthly and annual performance against the S&P500
Considering the annual return, **Bitcoin beat the S&P500 for the majority of the time**. Only from late 2018 to mid-2019, the S&P500 performed better than the cryptocurrency for a significant period of time. In total, the annual return of Bitcoin was higher than the S&P500's on **1874 out of 2617 days.**

For the monthly performance, the number of days (**1525/2617**) is a little lower, but Bitcoin still outperformed the S&P500 for the majority of the time. 

```python
# Insert all missing days into S&P500 data and fill the empty values with forward fill
sp500_daily = sp500.asfreq('D')
sp500_daily.fillna(method='ffill', inplace=True)

sp500_daily['daily_return'] = sp500_daily['close'].pct_change(1)
sp500_daily['monthly_return'] = sp500_daily['close'].pct_change(30)
sp500_daily['annual_return'] = sp500_daily['close'].pct_change(360)

# Monthly
ax_m = bitcoin.plot(y='monthly_return', label='Bitcoin', color=color_bitcoin)
sp500_daily.plot(y='monthly_return', figsize=(10, 7), ax=ax_m, grid=True, label='S&P500', title='Monthly returns of Bitcoin & the S&P500', xlabel='Date', color=color_sp500)

# Annual
ax_a = bitcoin.plot(y='annual_return', label='Bitcoin', color=color_bitcoin)
sp500_daily.plot(y='annual_return', figsize=(10, 7), ax=ax_a, grid=True, label='S&P500', title='Annual returns of Bitcoin & the S&P500', xlabel='Date', color=color_sp500) 

print('Number of days where Bitcoin\'s monthly return is higher than the S&P500\'s:', len(bitcoin.loc['2014-09-17':'2021-11-15'][bitcoin.loc['2014-09-17':'2021-11-15']['monthly_return'] > sp500_daily.loc['2014-09-17':'2021-11-15']['monthly_return']]), '/', len(bitcoin.loc['2014-09-17':'2021-11-15']))
print('Number of days where Bitcoin\'s annual return is higher than the S&P500\'s:', len(bitcoin.loc['2014-09-17':'2021-11-15'][bitcoin.loc['2014-09-17':'2021-11-15']['annual_return'] > sp500_daily.loc['2014-09-17':'2021-11-15']['annual_return']]), '/', len(bitcoin.loc['2014-09-17':'2021-11-15']))
```

## Monthly and annual performance against Gold
Bitcoin's performance compared to Gold is similar to comparison against the S&P500. Out of 2603 days, the monthly returns are higher at **1517 days, the annual at 1813 days.**

Compared against these two benchmarks, **Bitcoin wins in both cases.**

```python
# Insert all missing days into Gold data and fill the empty values with forward fill
daily_data = monthly_data.asfreq('D')
daily_data.fillna(method='ffill', inplace=True)

daily_data['daily_return'] = daily_data['gold_usd'].pct_change(1)
daily_data['monthly_return'] = daily_data['gold_usd'].pct_change(30)
daily_data['annual_return'] = daily_data['gold_usd'].pct_change(360)

# Monthly
ax_m = bitcoin.plot(y='monthly_return', label='Bitcoin', color=color_bitcoin)
daily_data.plot(y='monthly_return', figsize=(10, 7), ax=ax_m, grid=True, label='Gold', title='Monthly returns of Bitcoin & Gold', xlabel='Date', color=color_gold)

# Annual
ax_a = bitcoin.plot(y='annual_return', label='Bitcoin', color=color_bitcoin)
daily_data.plot(y='annual_return', figsize=(10, 7), ax=ax_a, grid=True, label='Gold', title='Annual returns of Bitcoin & Gold', xlabel='Date', color=color_gold) 

print('Number of days where Bitcoin\'s monthly return is higher than Gold\'s:', len(bitcoin.loc['2014-09-17':'2021-11-01'][bitcoin.loc['2014-09-17':'2021-11-01']['monthly_return'] > daily_data.loc['2014-09-17':'2021-11-01']['monthly_return']]), '/', len(bitcoin.loc['2014-09-17':'2021-11-01']))
print('Number of days where Bitcoin\'s annual return is higher than Gold\'s:', len(bitcoin.loc['2014-09-17':'2021-11-01'][bitcoin.loc['2014-09-17':'2021-11-01']['annual_return'] > daily_data.loc['2014-09-17':'2021-11-01']['annual_return']]), '/', len(bitcoin.loc['2014-09-17':'2021-11-01']))
```

# Correlation
## Correlation between the returns of the assets
Either daily, monthly or annual **returns of the three assets do not correlate.** This means, that high returns from one asset do not neccessarily mean high returns from another. 

```python
assets_data = pd.DataFrame()

# Bitcoin
assets_data['bitcoin_daily_return'] = bitcoin['daily_return']
assets_data['bitcoin_monthly_return'] = bitcoin['monthly_return']
assets_data['bitcoin_annual_return'] = bitcoin['annual_return']
assets_data['bitcoin_close'] = bitcoin['close']
assets_data['bitcoin_volume'] = bitcoin['volume']

# S&P500
assets_data['sp500_daily_return'] = sp500['close'].pct_change(1)
assets_data['sp500_monthly_return'] = sp500['close'].pct_change(30)
assets_data['sp500_annual_return'] = sp500['close'].pct_change(360)
assets_data['sp500_close'] = sp500['close']
assets_data['sp500_volume'] = sp500['volume']

# Gold
assets_data['gold_daily_return'] = daily_data['daily_return']
assets_data['gold_monthly_return'] = daily_data['monthly_return']
assets_data['gold_annual_return'] = daily_data['annual_return']
assets_data['gold_close'] = daily_data['gold_usd']

# Correlations & plots
daily_return_corr = assets_data[['bitcoin_daily_return', 'sp500_daily_return', 'gold_daily_return']].corr()
sns.heatmap(daily_return_corr, annot=True).set_title('Correlation between daily returns')
plt.show()
monthly_return_corr = assets_data[['bitcoin_monthly_return', 'sp500_monthly_return', 'gold_monthly_return']].corr()
sns.heatmap(monthly_return_corr, annot=True).set_title('Correlation between monthly returns')
plt.show()
annual_return_corr = assets_data[['bitcoin_annual_return', 'sp500_annual_return', 'gold_annual_return']].corr()
sns.heatmap(annual_return_corr, annot=True).set_title('Correlation between annual returns')
plt.show()
```

## Correlation between the prices of the assets
Whilst the returns of the assets have no connection, things look differently when considering the **close prices** of the assets. The **strongest correlation can be noticed between the S&P500 and Bitcoin** and between the S&P500 and Gold. Bitcoin's price does also correlate with the one of Gold, but not as strongly as it does with the S&P500. 

```python
# Correlations & plots
daily_return_corr = assets_data[['bitcoin_close', 'sp500_close', 'gold_close']].corr()
sns.heatmap(daily_return_corr, annot=True).set_title('Correlation between prices')
plt.show()
```

## Correlation between the price and the trading volume
#### Volume
It is widely known, that **Bitcoin is a young asset** with the first ever transaction happening in 2009. Bitcoin was not very popular throughout the first years of its existence, hence the **trading volume was low** and increased as Bitcoin gained popularity. The S&P500's trading volume on the other hand kept its level over time and oscillates around its mean at roughly **4 billion dollars a day.** 

```python
assets_data[['bitcoin_volume', 'sp500_volume']].plot(subplots=True, figsize=(10, 7), grid=True, xlabel='Date', title='Trading volume of Bitcoin and the S&P500 in $', color=[color_bitcoin, color_sp500])
```

#### Correlation between price and volume
For Bitcoin, there is a high correlation between the volume traded and the price, for the S&P500 on the other side this can not be said. The reason for the **high correlation for Bitcoin** is most likely the increasing popularity of the asset. As Bitcoin gained popularity (mostly for its high returns), money flew into the market and thus the price rose. On the other hand, there are also **periods with high losses**, where many people sell their Bitcoin(s), which can also be seen in trading volume data: the day with the by far highst volume was a day where the price went down. 

The high correlation get's even clearer when comparing trading volume and close price over time: **at times where Bitcoin's price peaked, also the trading volume was higher.** For the future, it is expected that the correlation will not be that strong anymore. Even though there is not enough data yet, the latest peak in November 2021 allows the assumption, that the volume is not so important for increasing prices anymore, as there does not seem to be a major increase in volume. Rather, **the volume seems to be stable now.** 

```python
price_vol_corr_bitcoin = assets_data[['bitcoin_close', 'bitcoin_volume']].corr()
price_vol_corr_sp500 = assets_data[['sp500_close', 'sp500_volume']].corr()

sns.heatmap(price_vol_corr_bitcoin, annot=True)
plt.show()
sns.heatmap(price_vol_corr_sp500, annot=True)
plt.show()
```

# Risk
## Broadness of returns
In history, Bitcoin gave much higher returns than the S&P500 (and also Gold), but this comes with a **high risk of losing money short-term**. Both the interquantile-range and the difference between the 90% and 10% quantiles are way higher for Bitcoin than they are for the S&P500. Only shortly after the COVID crisis hit in early 2020, the S&P500 became unstable. Considering a **long-term investment**, the uncertainty of daily returns does not matter that much. This will certainly only hold true for the future, if Bitcoin can keep performing better than the market. 

```python
rolling_bitcoin = assets_data['bitcoin_daily_return'].rolling(window=180)
quantiles_bitcoin = pd.DataFrame()
quantiles_bitcoin['q10'] = rolling_bitcoin.quantile(0.1)
quantiles_bitcoin['q25'] = rolling_bitcoin.quantile(0.25)
quantiles_bitcoin['q50'] = rolling_bitcoin.quantile(0.5)
quantiles_bitcoin['q75'] = rolling_bitcoin.quantile(0.75)
quantiles_bitcoin['q90'] = rolling_bitcoin.quantile(0.9)

quantiles_bitcoin.plot()
assets_data['bitcoin_daily_return'].plot(alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Daily return')
plt.title('Bitcoin daily return and quantiles')
```

```python
rolling_sp500 = sp500_daily['daily_return'].rolling(window=180)
quantiles_sp500 = pd.DataFrame()
quantiles_sp500['q10'] = rolling_sp500.quantile(0.1)
quantiles_sp500['q25'] = rolling_sp500.quantile(0.25)
quantiles_sp500['q50'] = rolling_sp500.quantile(0.5)
quantiles_sp500['q75'] = rolling_sp500.quantile(0.75)
quantiles_sp500['q90'] = rolling_sp500.quantile(0.9)

quantiles_sp500.plot()
assets_data['sp500_daily_return'].plot(alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Daily return')
plt.title('S&P500 daily return and quantiles')
```

```python
diff = pd.DataFrame()
diff['bitcoin_q90_10'] = quantiles_bitcoin['q90'].sub(quantiles_bitcoin['q10'])
diff['bitcoin_q75_25'] = quantiles_bitcoin['q75'].sub(quantiles_bitcoin['q25'])
diff['sp500_q90_10'] = quantiles_sp500['q90'].sub(quantiles_sp500['q10'])
diff['sp500_q75_25'] = quantiles_sp500['q75'].sub(quantiles_sp500['q25'])

ax = diff.plot(y=['bitcoin_q90_10', 'sp500_q90_10'], color=[color_bitcoin, color_sp500], figsize=(10, 7))
diff.plot(y=['bitcoin_q75_25', 'sp500_q75_25'], color=[color_bitcoin, color_sp500], linestyle='dashed', ax=ax)
plt.xlabel('Date')
plt.ylabel('Daily return difference')
plt.title('Difference between quantiles of daily return')
```

## Volatility
As already the broadness of returns suggested, the volatility for Bitcoin is way higher than it is for the S&P500. A **monthly volatility of 21.37%** or an **annual volatility of 74.56%** may make Bitcoin look like a bad investment compared to the **6.16%** and **21.47%** for the same volatility measures for the S&P500. Considering that Bitcoin yielded huge returns (**13079%**) over the time period (the S&P500 made roughly **134%**), the volatility does not look so bad anymore. The volatility of Bitcoin therefore meant gains most of the time which leads to good risk-adjusted returns.

Over the time, Bitcoin's volatility slightly increased. This is most likely because of the increase in trading volume and the money that floated into the market as a result. On the other hand, the monthly mean of daily returns also increased over time. It is expected that Bitcoin can keep its upward trend and also its high volatility. Even though it is a risky investment, such returns might be worth it. 

```python
# Total returns
bitcoin_total_return = (assets_data['bitcoin_close'][-1] - assets_data['bitcoin_close'][0])/assets_data['bitcoin_close'][0]
sp500_total_return = (assets_data['sp500_close'].dropna()[-1] - assets_data['sp500_close'][0])/assets_data['sp500_close'][0]

print('Total return Bitcoin: ', bitcoin_total_return * 100, '%')
print('Total return S&P500: ', sp500_total_return * 100, '%', '\n')

# Annualized returns
months = int((assets_data.index[-1] - assets_data.index[0])/np.timedelta64(1, 'M'))
annualized_return_bitcoin = ((1 + bitcoin_total_return) ** (1/7)) - 1
annualized_return_sp500 = ((1 + sp500_total_return) ** (1/7)) - 1

# Calculate annualized volatility from the standard deviation
monthly_volatility_bitcoin = (assets_data['bitcoin_daily_return'].std() * np.sqrt(30))
monthly_volatility_sp500 = (assets_data['sp500_daily_return'].std() * np.sqrt(30))
annual_volatility_bitcoin_total = assets_data['bitcoin_daily_return'].std() * np.sqrt(365)
annual_volatility_sp500_total = (assets_data['sp500_daily_return'].std() * np.sqrt(365))

print('Annualized volatility Bitcoin (one month): ', monthly_volatility_bitcoin)
print('Annualized volatility Bitcoin (one year): ', annual_volatility_bitcoin_total, '\n')
print('Annualized volatility S&P500 (one month): ', monthly_volatility_sp500)
print('Annualized volatility S&P500 (one year): ', annual_volatility_sp500_total)
```

```python
monthly_volatilities_bitcoin = []
monthly_volatilities_sp500 = []
annual_volatilities_bitcoin = []
annual_volatilities_sp500 = []

for year in range(2014, 2022):
    # Volatility
    year_str = str(year)
    monthly_volatility_bitcoin = (assets_data['bitcoin_daily_return'][year_str].std() * np.sqrt(30))
    monthly_volatility_sp500 = (assets_data['sp500_daily_return'][year_str].std() * np.sqrt(30))
    annual_volatility_bitcoin = assets_data['bitcoin_daily_return'][year_str].std() * np.sqrt(365)
    annual_volatility_sp500 = (assets_data['sp500_daily_return'][year_str].std() * np.sqrt(365))
    
    monthly_volatilities_bitcoin.append(monthly_volatility_bitcoin)
    monthly_volatilities_sp500.append(monthly_volatility_sp500)
    annual_volatilities_bitcoin.append(annual_volatility_bitcoin)
    annual_volatilities_sp500.append(annual_volatility_sp500)

plt.plot(range(2014, 2022), monthly_volatilities_bitcoin, label='Bitcoin monthly volatility', color=color_bitcoin, linestyle='dashed')
plt.plot(range(2014, 2022), monthly_volatilities_sp500, label='S&P500 monthly volatility', color=color_sp500, linestyle='dashed')
plt.plot(range(2014, 2022), annual_volatilities_bitcoin, label='Bitcoin annual volatility', color=color_bitcoin)
plt.plot(range(2014, 2022), annual_volatilities_sp500, label='S&P500 annual volatility', color=color_sp500)
plt.legend()
plt.title('Bitcoin and S&P500 volatility')
```

```python
mean_daily_returns = pd.DataFrame()
mean_daily_returns['mean_bitcoin'] = assets_data['bitcoin_daily_return'].resample('M').mean()
mean_daily_returns['days_from_start'] = (mean_daily_returns.index - mean_daily_returns.index[0]).days

x = mean_daily_returns['days_from_start'].values.reshape(-1, 1)
y = mean_daily_returns['mean_bitcoin'].values

model = LinearRegression().fit(x, y)
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1, normalize=False)
mean_daily_returns['prediction'] = model.predict(x)

ax = mean_daily_returns.plot(y='mean_bitcoin', color=color_bitcoin, label='Bitcoin mean monthly return')
mean_daily_returns.plot(y='prediction', color='black', linestyle='dashed', ax=ax, linewidth=1, label='Trendline')
plt.xlabel('Date')
plt.ylabel('Mean monthly return')
plt.title('Change of mean monthly return over time')
```

## Risk adjusted returns
#### Sharpe ratio
It was discussed above that Bitcoin's high volatility comes with even higher returns. If the returns are high enough, a good Sharpe ratio will be the result. It turns out that for Bitcoin the Sharpe ratio is indeed good. With **1.35** it is above 1, which is the threshold for a Sharpe ratio to be considered good and also above the ratio for the S&P500 (**0.6**). According to the Sharpe ratio, **Bitcoin is worth an investment.**

```python
# Sharpe ratio
risk_free_rate = 0.0
sharpe_ratio_bitcoin = (annualized_return_bitcoin - risk_free_rate)/annual_volatility_bitcoin_total
sharpe_ratio_sp500 = (annualized_return_sp500 - risk_free_rate)/annual_volatility_sp500_total

print('Sharpe ratio Bitcoin (one year): ', sharpe_ratio_bitcoin)
print('Sharpe ratio S&P500 (one year): ', sharpe_ratio_sp500)
```

#### Sortino ratio 
When only considering the downside risk (i.e., the standard deviation of all the past losses), Bitcoin still performs better than the S&P500. 

```python
# Downside returns
downside_returns_bitcoin = assets_data[assets_data['bitcoin_daily_return'] < 0]['bitcoin_daily_return']
downside_returns_sp500 = assets_data[assets_data['sp500_daily_return'] < 0]['sp500_daily_return']

# Expected returns
expected_return_bitcoin = assets_data['bitcoin_daily_return'].mean()
expected_return_sp500 = assets_data['sp500_daily_return'].mean()

# Downside variation
downside_std_bitcoin = downside_returns_bitcoin.std()
downside_std_sp500 = downside_returns_sp500.std()

# Calculate the sortino ratio
sortino_ratio_bitcoin = (expected_return_bitcoin - risk_free_rate)/downside_std_bitcoin
sortino_ratio_sp500 = (expected_return_sp500 - risk_free_rate)/downside_std_sp500

# Print the results
print("Expected return Bitcoin: ", expected_return_bitcoin * 100)
print("Expected return S&P500: ", expected_return_sp500 * 100, '\n')
print("Downside risk Bitcoin: ", downside_std_bitcoin * 100)
print("Downside risk S&P500: ", downside_std_sp500 * 100, '\n')
print("Sortino ratio Bitcoin: ", sortino_ratio_bitcoin)
print("Sortino ratio S&P500: ", sortino_ratio_sp500, '\n')
```

# Inflation
## Inflation rates
Over the period, **the Consumer Price Index constantly growed**, at some times more than at others. Espacially after the beginning of the COVID crisis in 2020 annual inflation started to grow up to a staggering 6.22% which is far bigger than the average annual inflation rate of around 2%. Also the monthly inflation rate saw a rise lately as it reached its maximum at almost 1% whilst before that being oftentimes even negative.

```python
# Monthly inflation rate
monthly_data['monthly_inflation_rate'] = monthly_data['cpi_us'].pct_change(1)

# Annual inflation rate
monthly_data['annual_inflation_rate'] = monthly_data['cpi_us'].pct_change(12)

# Plot the data
monthly_data.plot(y=['annual_inflation_rate', 'monthly_inflation_rate'], grid=True, color=['blue', 'red'], label=['Annual inflation rate', 'Monthly inflation rate'])
plt.xlabel('Date')
plt.ylabel('Inflation rate')
plt.title('Inflation rate over time')

monthly_data[['monthly_inflation_rate', 'annual_inflation_rate']].describe()
```

## Assets against inflation
Had one invested money in any of the assets at the first day of a month and compared the return at the last day of the month to the change in the Consumer Price Index (=inflation), Bitcoin would have beaten the monthly inflation the most times out of the 3 assets. However, the lead is very thin with the S&P500 coming second being only one month below Bitcoin. Altogether, all three of the assets beat the inflation roughly **50-60%** of the time. This fact is neither bad nor good and furthermore less important than the comparison of the yearly return to the annual inflation. All three assets are good hedges against inflation: the S&P500 and Gold both yielded higher returns than the inflation rate was in 4 out of 6 years, whilst Bitcoin could even improve on this and yield higher returns in 5 out of 6 years. 

**Bitcoin can therefore be considered a good hedge against inflation.**

# Portfolio composition and conclusion
The last chapters showed that Bitcoin yields high returns but also has a high volatility. To reduce the volatility in a portfolio consisting of Bitcoin, an Index fund of the S&P500 & Gold, Bitcoin should only be a tiny part of the portfolio. For **minimum volatility**, the following asset allocation is recommended: 

- **Bitcoin:** 1.07%
- **S&P500:** 25.57%
- **Gold:** 73.36%

Bitcoin showed a clear upward trend in the time period. It is expected that it can keep this trend alive in the future. Therefore, it is better to go for a portfolio with high risk-adjusted returns but as a consequence higher volatility. Considering the Sharpe ratio as a measure for the risk-adjusted returns, the **following asset allocation is recommended:** 

- **Bitcoin:** 23.57%
- **S&P500:** 34.34%
- **Gold:** 42.09%

```python
monthly_returns = pd.DataFrame()
monthly_returns['bitcoin'] = assets_data['bitcoin_daily_return']
monthly_returns['sp500'] = assets_data['sp500_daily_return']
monthly_returns['gold'] = assets_data['gold_daily_return']

# Annualized mean returns
mus = (1 + monthly_returns.mean()) ** 252 - 1

# Covariances and variances
cov = monthly_returns.cov() * 252
```

```python
# Number of assets
n_assets = 3
# Number of randomly generated portfolios
n_portfolios = 10000

# Store mean-variance pairs for plotting
mean_variance_pairs = []

# Minimum volatility and the corresponding weights
min_volatility = np.infty
min_volatility_weights = None

# Maximum Sharpe ratio and the corresponding weights
max_sharpe_ratio = -np.infty
max_sharpe_ratio_weights = None

# Minimum Sharpe ratio
min_sharpe_ratio = np.infty

# Assets
assets = list(monthly_returns.columns)

np.random.seed(42)
risk_free_rate = 0

# Loop through and generate random portfolios
for _ in range(n_portfolios):
    # Random weights
    weights = np.random.rand(n_assets)
    # Ensure weights to sum to 1
    weights = weights/sum(weights)

    # Loop over asset pairs and compute portfolio return and variance
    expected_variance = 0
    expected_return = 0
    for i in range(len(assets)):
        expected_return += weights[i] * mus.loc[assets[i]]
        for j in range(len(assets)):
            # Add variance/covariance for each asset pair
            expected_variance += weights[i] * weights[j] * cov.loc[assets[i], assets[j]]
    
    # Save the minimum volatility and the corresponding weights
    if expected_variance ** 0.5 < min_volatility:
        min_volatility = expected_variance ** 0.5
        min_volatility_weights = weights
	
    # Get the Sharpe ratio and save it and the corresponding weights if its maximal (same for minimum)
    sharpe_ratio = (expected_return - risk_free_rate)/(expected_variance ** 0.5)
    if sharpe_ratio > max_sharpe_ratio:
        max_sharpe_ratio = sharpe_ratio
        max_sharpe_ratio_weights = weights
    elif sharpe_ratio < min_sharpe_ratio:
        min_sharpe_ratio = sharpe_ratio
            
    # Store return, variance and Sharpe ratio
    mean_variance_pairs.append([expected_return, expected_variance, sharpe_ratio])
```

```python
print('Minimum volatility: ', min_volatility, '\nMaximum Sharpe ratio: ', max_sharpe_ratio, '\n')

print('Asset allocation at minimum volatility:')
for a, w in zip(assets, min_volatility_weights):
    print(a, ':', w * 100, '%')

print('\nAsset allocation at maximum Sharpe ratio:')
for a, w in zip(assets, max_sharpe_ratio_weights):
    print(a, ':', w * 100, '%')
```

```python
# Plot the portfolios
mean_variance_pairs = np.array(mean_variance_pairs)

norm = plt.Normalize(min_sharpe_ratio, max_sharpe_ratio)
sm = plt.cm.ScalarMappable(cmap="RdBu", norm=norm)
sm.set_array([])

fig = sns.scatterplot(x=mean_variance_pairs[:,1] ** 0.5, y=mean_variance_pairs[:,0], palette='RdBu', hue=mean_variance_pairs[:,2])
plt.xlabel('Annualized Risk (Volatility)')
plt.ylabel('Annualized Return')
plt.title('Sample of Random Portfolios')
plt.grid()

fig.get_legend().remove()
fig.figure.colorbar(sm)
```
